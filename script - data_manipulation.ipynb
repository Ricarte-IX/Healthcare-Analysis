```python
# import library
import pandas as pd
```

```python
# read dataset
dataset = pd.read_csv('raw_healthcare_dataset.csv')
dataset
```

```python
# make a copy of the dataset
dataset_modified = dataset.copy()
dataset_modified['Name']

# standardize the column 'Name'
dataset_modified['Name'] = dataset_modified['Name'].str.title()
dataset_modified
```

```python
# change the data type of columns 'Date of Admission' and 'Discharge Date' to datetime
dataset_modified['Date of Admission'] = pd.to_datetime(dataset_modified['Date of Admission'], dayfirst=True)
dataset_modified['Discharge Date'] = pd.to_datetime(dataset_modified['Discharge Date'], dayfirst=True)

# extract the date part only
dataset_modified['Date of Admission'] = dataset_modified['Date of Admission'].dt.date
dataset_modified['Discharge Date'] = dataset_modified['Discharge Date'].dt.date

print(dataset_modified.dtypes)
```

```python
# save the cleaned dataset as excel file
dataset_modified.to_excel('cleaned_healthcare_dataset.xlsx', sheet_name='Healthcare Data', index=False)
```
